# **LIMPIEZA DE DATOS**

1. **Tabla Candidate_status_update:**
Contiene todas las respuestas de los estudiantes preseleccionados, así como las respuestas de los reclutadores después de que los estudiantes hayan declarado su interés en la oferta.

```python
from google.colab import auth
import pandas as pd

# get credentials
auth.authenticate_user()

query = '''
SELECT *
FROM `data-analytics-bootcamp-363212.jobteaser.candidate_status_update`
'''
df_candidate = pd.read_gbq(query, project_id="data-analytics-bootcamp-363212")
df_candidate.head()
```

2. **Tabla Optin:** Contiene toda la información de inclusión y exclusión voluntaria de los estudiantes. Este es un interruptor para activar "Buscando oportunidades" y, por lo tanto, aceptamos recibir ofertas de empresas a través del producto Shortlist.

```python
# Tabla Optin
query2 = '''
SELECT *
FROM `data-analytics-bootcamp-363212.jobteaser.optin`
'''
df_optin = pd.read_gbq(query2, project_id="data-analytics-bootcamp-363212")
df_optin.head()
```

3.**Tabla dim_schools:** Contiene toda la información sobre las escuelas asociadas a JT.

```python
# Tabla dim_school
query3 = '''
SELECT *
FROM `data-analytics-bootcamp-363212.jobteaser.dim_schools`
'''
df_school = pd.read_gbq(query3, project_id="data-analytics-bootcamp-363212")
df_school.head()
```
